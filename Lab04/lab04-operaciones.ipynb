# Matrix operations

(You can take Python4Examples.py as a reference) You must print on console every change or operation performed.

1. Create 2 matrices A and B of 15x15 each with random inputs between -10 and 10.

```python
import numpy as np

a = np.random.randint(-10, 10,  size=(15, 15))
b = np.random.randint(-10, 10,  size=(15, 15))
print("A = \n" + str(a))
print("B = \n" + str(b))
```

2. Change the diagonals of both matrices to different values using np.fill_diagonal.

```python
np.fill_diagonal(a, 0)
np.fill_diagonal(b, 0)

print("A = \n" + str(a))
print("B = \n" + str(b))
```

3. Make the sum, difference, product between A and B.

```python
print("A + B = \n" + str(a+b))
print("A - B = \n" + str(a-b))
print("A * B = \n" + str(a*b))
```

4. Find the transpose of A and B.

```python
print("A^T = \n " + str(a.transpose()))
print("B^T = \n " + str(b.transpose()))
```

5. Create two vectors c and d (1 x 20) with random inputs using np.random.

```python
c = np.random.randint(10,  size=(1, 20))
d = np.random.randint(10,  size=(1, 20))

print("C =  " + str(c))
print("D =  " + str(d))
```

6. Make the product point between c and d.

```python
d = d.transpose()

cd = np.dot(c,d)
print("C * D = " + str(cd))
```
